```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X = data.drop(columns=['energy_consum_per_cap_mil_btu'])
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['variable'] = X.columns

#view VIF for each explanatory variable 
vif
```

None of the VIFs are above 5, so there is not too much multicollinearity. Regardless, we will use a ridge regression model because we have so many features for so few data points. We will also standardize all of the feature data so we can compare the coefficients to see which have the largest affect on energy usage.

```python
mat = np.asarray(X)
mean = np.mean(mat, axis = 0)
sd = np.std(mat, axis = 0)
scaled = (mat - mean) / sd
```

```python
# add offset
scaled = np.hstack((scaled, np.ones(len(scaled)).reshape(-1,1)))
```

We have a small number of datapoints that vary a lot, and we only want to know the importance of features, not predict energy consumption from new state data (there are only 51 states in the USA and we have data for all of them). Therefore, I will not create a test set and will use Leave-One-Out Cross Validation to ensure the error of the model is low. This way, we can incorporate all of the data into finding the most important features.

```python
from sklearn.model_selection import KFold
import statsmodels.api as sm
```

```python
def MSE(y, pred):
    return np.sum((y-pred)**2)/len(y)

def ridge(train_y, test_y, train_x, test_x, lam):
    ols_model = sm.regression.linear_model.OLS(train_y, train_x)
    results = ols_model.fit_regularized(method='elastic_net', alpha=lam, L1_wt=0)
            
    w = np.array(results.params).reshape([len(results.params),1])

    train_pred = np.matmul(train_x,w)
    test_pred = np.matmul(test_x,w)

    train_MSE = MSE(train_y, train_pred.flatten())
    test_MSE = MSE(test_y, test_pred.flatten())
    
    return train_MSE, test_MSE, w, test_pred

def ridge_lam(lambdavals, X, y, k):
    cv = KFold(n_splits = k)
    for lam in lambdavals:
        avg_train_MSE = 0
        avg_val_MSE = 0

        for train_index, val_index in cv.split(X):
            X_traink, X_val = X[train_index], X[val_index]
            y_traink, y_val = y[train_index], y[val_index]
            train_MSE, val_MSE, w, pred = ridge(y_traink, y_val, X_traink, X_val, lam)
            avg_train_MSE += train_MSE
            avg_val_MSE += val_MSE

        avg_train_MSE = avg_train_MSE/k
        avg_val_MSE = avg_val_MSE/k

        print(f'lambda: {lam}, avg_train_MSE: {avg_train_MSE}, avg_val_MSE: {avg_val_MSE}')
```

```python
lambdas = [0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100]
ridge_lam(lambdas, scaled, data['energy_consum_per_cap_mil_btu'], 51)
```

```python
lambdas = np.array(list(range(1,11)))/1000
ridge_lam(lambdas, scaled, data['energy_consum_per_cap_mil_btu'], 51)
```

Through telescopic search, it looks like a lambda of 0.002 got the lowest validation error and nearly the lowest training error. Therefore I will use that for getting our coefficients.

```python
ols_model = sm.regression.linear_model.OLS(data['energy_consum_per_cap_mil_btu'], X)
results = ols_model.fit_regularized(method='elastic_net', alpha=0.002, L1_wt=0)
w = np.array(results.params).reshape([len(results.params),1])
```
